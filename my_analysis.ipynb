```python
import pandas as pd


df = pd.read_csv('Average_Screen_Time_and_Usage_by_Country.csv')
```

```python
# --- שלב 3: עיבוד הנתונים (Data Processing) ---

# 1. הצגת מידע כללי על הנתונים (כמה ערכים חסרים יש?)
print("סיכום ערכים חסרים לפני ניקוי:")
print(df.isnull().sum())

# 2. ניקוי ערכים חסרים (הסרת שורות ריקות)
df_cleaned = df.dropna().copy()

# 3. הסרת כפילויות (אם קיימות)
df_cleaned = df_cleaned.drop_duplicates()


df_cleaned.columns = [
    'Country', 'Total_Time', 'Mobile_Time', 'PC_Time', 'Internet_Users_Pct',
    'Mobile_Speed', 'Fixed_Speed', 'Streaming_TV_Pct', 'Gaming_Pct',
    'Console_Time', 'Social_Media_Time', 'Social_Accounts', 'Podcast_Pct',
    'Podcast_Time', 'Music_Streaming_Pct', 'Music_Time'
]

print(df_cleaned.head())




df_cleaned.to_csv('cleaned_screen_time_data.csv', index=False)
```
